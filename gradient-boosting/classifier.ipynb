```python
from pandas import read_csv

data = read_csv('gbm-data.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = [row[1:1777] for row in data.values]
y = [row[0] for row in data.values]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=241)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from numpy import array
from math import exp
from sklearn.metrics import log_loss


def applySDF(X, y, clf):
    min_metric = 123123123
    min_stage = 0
    
    for stage, pred in enumerate(clf.staged_decision_function(X)):
        sigma_pred = [1/(1 + exp(-elem)) for elem in pred]
        metric = log_loss(y, sigma_pred)
        if metric < min_metric:
            min_metric = metric
            min_stage = stage
     
    print(min_metric, min_stage) 

learning_rate_array = [1, 0.5, 0.3, 0.2, 0.1]

for rate in learning_rate_array:
    gbclf = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=rate)
    gbclf.fit(X_train,y_train)
    applySDF(X_train, y_train, gbclf)
    applySDF(X_test, y_test, gbclf)      
# 0.00020857611394801468 180
# 0.5822942594278475 0

# 0.001022368076368321 249
# 0.5560965876181979 6

# 0.006887291001447366 249
# 0.5432596932316428 10

# 0.024906124380084924 249
# 0.5314949511258739 36

# 0.08936962416274086 249
# 0.5254924770369842 51
```
